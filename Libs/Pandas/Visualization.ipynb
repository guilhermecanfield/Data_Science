# Creating Bar Graphs to Show Counts

```python
def rank_performance(stock_price):
    if stock_price <= 10:
        return 'Poor'
    elif stock_price <= 50:
        return 'Satisfactory'
    else:
        return 'Stellar'
```

```python
bb['Close'].apply(rank_performance).value_counts(normalize = True).plot(kind = 'pie')
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind = 'bar')
```

```python
bb['Close'].apply(rank_performance).value_counts().plot(kind = 'barh')
```

```python
def above_below(stock_price):
    if stock_price >= bb['Close'].mean():
        return 'Above Avarage'
    else:
        return 'Below Avarage'
```

```python
plt.style.use('dark_background')
bb['Close'].apply(above_below).value_counts().plot(kind = 'pie')
```
