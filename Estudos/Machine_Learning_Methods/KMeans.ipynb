```python
X = pd.get_dummies(df)
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_X = scaler.fit_transform(X)

from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)

cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
X['cluster'] = cluster_labels
X
```

```python
plt.figure(figsize=(12,6),dpi=200)
X.corr()['cluster'].iloc[:-1].sort_values().plot(kind='bar');
```

```python
ssd = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    
    ssd.append(model.inertia_) # SSD point --> cluster center
```

```python
plt.plot(range(2,10),ssd,'o--');
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
```

# K-Means Clustering for Color Quantization

```python
import matplotlib.image as mpimg
```

```python
image = mpimg.imread('../DATA/palm_trees.jpg')
```

```python
image.shape #Height, width, colour_channels
```

```python
image #Each column from the array represent one amount of colour (Red|Green|Blue)
```

```python
plt.figure(dpi=150)
plt.imshow(image);
```

```python
# Transform the 3d array into a 2d array
# (H,W,C) ---> (H*W,C)
(h,w,c) = image.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
image_2d = image.reshape(h*w,c)
image_2d #notice that the number of bruckets fall to two
```

```python
image_2d.shape
```

```python
model = KMeans(n_clusters=6)

labels = model.fit_predict(image_2d)
```

```python
model.cluster_centers_
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
rgb_codes
```

```python
labels
```

```python
rgb_codes[labels]
```

```python
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
quantized_image
```

```python
plt.figure(dpi=200)
plt.imshow(quantized_image);
```
