```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,9)
plt.ylim(0,18)
plt.title('String Title')
plt.savefig('myfirstplot.png')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.5,0.5]) # Here I can control de dimension of the plot on the figure
axes.plot(x,y)
```

```python
# Creating the data that will be ploted
a = np.linspace(0,10,11)
b = a ** 4
x = np.arange(0,10)
y = 2 * x
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([0,0.5,0.5,0.5]) # Every positional arguments here must be between 0 and 1 
# [x the closer to one, the more to the right the figure
#  y the closer to one, the higher level of the figure
#  width
#  height]
axes2.plot(a,b)
axes2.set_xlim(0,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('zoomed In')
```

```python
fig = plt.figure(figsize = [10,1], dpi = 200) # figsize = [width, height], dpi = dots per inch = pontos por polegada

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight') # Include the x and y labels in the png figure
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1)

for ax in axes:
    ax.plot(a,b)
    
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(10,5), dpi=100)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)

axes[1][1].plot(x,y)

fig.suptitle('Figure Level', fontsize=20)

fig.subplots_adjust(wspace=0.5, hspace=0.5)

fig.savefig('new_subplots.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, label='X vs X')

ax.plot(x,x**2, label='X vs X^2')

# ax.legend(loc='upper right')
ax.legend(loc=(1.05,0.5)) # Determine the "cordinates" of de legend. X, Y - 0 left | 1 right 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,x, color='#faf205', label='X vs X', lw=10, ls='--') # RGB(Red, Green, Blue) HEX(Hexadecimal) Code it's a formula how much Red Green and Blue 
# the colour will countains. Search for hex color picker in google to discover the exactly code for a color

lines = ax.plot(x,x+1,color='purple', label='X vs X+1', linewidth=5, linestyle=':')
lines[0].set_dashes([1,2,1,2,10,2]) #1 point, 2 spaces, 1 point, 2 spaces, 10 points, 2 spaces
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color='#faf205', label='X vs X', lw=5, marker='+', ms=20) # RGB(Red, Green, Blue) HEX(Hexadecimal) Code it's a formula how much Red Green and Blue 
# the colour will countains. Search for hex color picker in google to discover the exactly code for a color

ax.plot(x,x+1,color='purple', label='X vs X+1', linewidth=2, marker='o', markersize=10,
        markerfacecolor='red', markeredgewidth=4 ,markeredgecolor='orange')

ax.legend()
```
